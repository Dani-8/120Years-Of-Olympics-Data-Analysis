```python
plt.scatter("Height","Weight",data=df,color="darkblue")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height" , y="Weight" , data=df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelets Weight vs Height")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height" , y="Weight" , data=df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelets Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height" , y="Weight" , hue="Sex",data=df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelets Weight vs Height")
plt.show()
```

```python
 df["Medal"].unique()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height" , y="Weight" , hue="Sex",style="Medal",data=df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelets Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height" , y="Weight" , hue="Sex",style="Medal",size="Age",data=df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelets Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height" , y="Weight" , hue="Sex",data=df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelets Weight vs Height")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height" , hue="Sex" , data=df)
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height" , hue="Sex" , data=df , kind="kde")
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.barplot(x="Medal" , y="Height", hue="Sex" , data=df)
plt.title("Medals by Height")
plt.show()
```

```python
sns.catplot(x="Medal" , y="Height", hue="Sex", col="Season", data=df)
plt.show()
```

```python
sns.catplot(x="Medal" , y="Height", hue="Sex", col="Season", kind="bar",data=df)
plt.show()
```
